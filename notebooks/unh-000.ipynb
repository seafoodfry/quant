```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Data comes from
# https://www.unitedhealthgroup.com/investors/stock.html
dividend_records = [
    # 2025
    (pd.Timestamp('2025-09-15'), 2.21),
    (pd.Timestamp('2025-06-16'), 2.21),
    (pd.Timestamp('2025-03-10'), 2.10),
    # 2024
    (pd.Timestamp('2024-12-09'), 2.10),
    (pd.Timestamp('2024-09-16'), 2.10),
    (pd.Timestamp('2024-06-17'), 2.10),
    (pd.Timestamp('2024-03-11'), 1.88),
    # 2023
]
```

```python
df = pd.DataFrame(dividend_records, columns=['date', 'dividend_amount'])
df = df.sort_values('date').reset_index(drop=True)
```

```python
print("UNH Dividend Data:")
print(df)
print(f"\nTotal dividends: ${df['dividend_amount'].sum():.2f}")
print(f"Average dividend: ${df['dividend_amount'].mean():.2f}")
```

```python
df['dividend_growth'] = df['dividend_amount'].pct_change() * 100
```

```python
fig, ax1 = plt.subplots(figsize=(8, 5))

# Left axis - dividend amount.
ax1.plot(df['date'], df['dividend_amount'], marker='o', linewidth=1, markersize=3, color='steelblue', label='Dividend Amount')
ax1.set_xlabel('Date')
ax1.set_ylabel('Dividend Amount ($)', color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue')
ax1.grid(True, alpha=0.3)

# Right axis - dividend growth.
ax2 = ax1.twinx()
ax2.plot(df['date'], df['dividend_growth'], marker='s', linewidth=1, markersize=3, color='green', label='Growth Rate')
ax2.set_ylabel('Dividend Growth (%)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title('UNH Dividend Amount & Growth Rate')
#plt.xticks(rotation=45)
ax1.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
