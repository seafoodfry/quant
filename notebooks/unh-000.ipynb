```python
df = pd.DataFrame(dividend_records, columns=['exdividend_date', 'dividend_amount'])
df = df.sort_values('exdividend_date').reset_index(drop=True)
```

```python
df['dividend_growth'] = df['dividend_amount'].pct_change() * 100
```

```python
print(f"\nTotal dividends: ${df['dividend_amount'].sum():.2f}")
print(f"Average dividend: ${df['dividend_amount'].mean():.2f}")

df.head()
```

```python
fig, ax1 = plt.subplots(figsize=(8, 5))

# Left axis - dividend amount.
ax1.plot(df['exdividend_date'], df['dividend_amount'], marker='o', linewidth=1, markersize=3, color='steelblue', label='Dividend Amount')
ax1.set_xlabel('Date')
ax1.set_ylabel('Dividend Amount ($)', color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue')
ax1.grid(True, alpha=0.3)

# Right axis - dividend growth.
ax2 = ax1.twinx()
ax2.plot(df['exdividend_date'], df['dividend_growth'], marker='s', linewidth=1, markersize=3, color='green', label='Growth Rate')
ax2.set_ylabel('Dividend Growth (%)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title('UNH Dividend Amount & Growth Rate')
#plt.xticks(rotation=45)
ax1.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
